```python
#Vending machine with python.
#creating two dictionaries that hold the names and costs of each item, one for drinks and the other for snacks. 
#The stock dictionary is then defined, keeping track of the quantity of each item in stock.

# a seperate dictionary for snack and drinks
snacks = {
    "A": {"name": "Lays", "price": 6.00,}, 
    "B": {"name": "Popcorn", "price": 5.00},
    "C": {"name": "Chettos", "price": 12.00},
    "D": {"name": "Takis", "price": 11.00},
    "E": {"name": "Mars", "price": 3.00}
}

drinks = {
    "F": {"name": "Fanta", "price": 2.00},
    "G": {"name": "Juice", "price": 2.50},
    "H": {"name": "Water", "price": 1.50},
    "I": {"name": "Pepsi", "price": 5.50},
    "J": {"name": "Coke 0", "price": 6.00}
}

# Dictionary for stocks (Item may run out of stock)
stock = {
    "A": 1, #If you wanna try on testing this out try A/E for 2 times(make sure to pay!)
    "B": 4,
    "C": 3,
    "D": 3,
    "E": 1,
    "F": 5,
    "G": 8,
    "H": 2,
    "I": 6,
    "J": 3
}


#Defining function that i can use later on ------------------------------
def check_stock(item): # Function to check if item is in stock
    if stock[item] > 0: # If in stock
        return True # If true "return" makes the program to continue running the main program after a function has been exited.
    else:
        return False 

#Then if i stock/not this if/else statement will show
def dispense(item): #dispenses the item and updates the stock if the item is in stock
    if check_stock(item): #If in stock
        print("Dispensing " + snacks[item]["name"] + "..." if item in snacks else "Dispensing " + drinks[item]["name"] + "...")
        stock[item] -= 1 # Value in stock will minus
    else:
        print("Item is out of stock.") 

#A way to manage users money
def calculate_change(cost, money): # Function to calculate change
    change = money - cost # Formula to result into change
    if change <= 0: #if "money" inputed by the user is less than or equal to 0
        print("Insufficient funds. Please try again.")
    else:
        print("Collect your  $" + str(change) + " change.")
    return change        

# Function to display the snack or drink options
def display_options(options):
    for key, value in options.items():
        print(f'{key}: {value["name"]} - ${value["price"]}')

# Defining a Receipt       
def generate_receipt(item, cost, change):
    print("\n\n------------RECEIPT------------")
    print("Item: " + item)
    print("Cost: $" + str(cost))
    print("Inserted cash: $" + str( money))
    print("Change: $" + str(change))
    print("Thank you for your purchase!")
    print("-------------------------------\n\n")  
      
#while loop that enables the user to choose between drinks, snacks, quit.
#Then displays the options, picks an item, inputs money, and prints a receipt.
#It will print a notice to that effect if the user makes an invalid selection or if the item is out of stock.
#-------------------------------------------------------------------------------
# Main loop
while True:
    print("\n- - - Welcome to the Vending Machine! - - -")#My introduction

    print("\nPlease select an option:")#Printing drink or snack option or quit!(User selection)
    print("(S)nacks")
    print("(D)rinks")
    print("(Q)uit")
    
    
    choice = input() #getting users input (S/D/Q)
    
    #S/D/Q IF-ELIF-ELSE
    #If Choice = S (Show snacks options)
    if choice == "S": 
        display_options(snacks)
       # Ask user for snack selection
        item = input("\nPlease select a snack option: ").upper()
        # Check if item is valid
        if item in snacks:
                cost = snacks[item]["price"]
                # Check if item is in stock
                if check_stock(item): 
                    # Ask user for money
                    money = int(input("\nGreat your balance is$"+ str(cost) +"." + "Please insert cash:"))
                    # Calculating the change and dispensing the item
                    change = calculate_change(cost, money)
                    if money >= snacks[item]["price"]:
                     dispense(item)
                     generate_receipt(snacks[item]["name"], cost, change) 
                #Not in-stock               
                else: 
                    print("Item is out of stock.") 
        # Not in "item in snacks"            
        else:
         print("Invalid selection.")
    
    #If Choice = D (Show drinks options)       
    # Show drink options (Same method)
    elif choice == "D":
        display_options(drinks)
       # Ask user for drink selection
        item = input("\nPlease select a snack option: ").upper()
        # Check if item is valid
        if item in drinks:
                cost = drinks[item]["price"]
                # Check if item is in stock
                if check_stock(item):
                    # Ask user for money
                    money = int(input("\nGreat your balance is $"+ str(cost) +"." + "Please insert cash:"))
                    # Calculating the change and dispensing the item
                    change = calculate_change(cost, money)
                    if money >= drinks[item]["price"]:
                     dispense(item)
                     generate_receipt(drinks[item]["name"], cost, change)
                #Not in-stock                      
                else:
                    print("Item is out of stock.")
        # Not in "item in drinks"                
        else:
         print("Invalid selection.")

#Elif Choice = Q (break)       
    elif choice == "Q":
      print('Thank you for using the vending machine!')
      break
#Else Choice(not S/D/Q) = Invalid
    else:
      print("Invalid selection.")
 
# - - -- - - - -- - -- - - -- - --  Main loop ends - - - -- - -- - - -- - - -- -
#If user wishes to continue! if-else statement
# Exit the program
    add = input("Would you like to continue? (Y) to continue or ANY key to QUIT: ").upper()
    if add == "Y":
     continue #Back to main loop
    else:
     print("Thank you for using the vending machine. Have a great day!")
     break #Quit program
     
# the end!
```
